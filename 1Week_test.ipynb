```python
import numpy as np
A = input()
B = input()

A = A.split(' ')
B = B.split(' ')

A = list(map(int, A))
B = list(map(int, B))

X = np.abs(A[0] - B[0])
Y = np.abs(A[1] - B[1])

if (X == 0) and (Y == 0):
    print('DOT')

elif (X == 0) or (Y == 0):
    print('SEGMENT')
    
elif X == Y:
    print('SQUARE')
    
elif X != Y:
    print('RECTANGLE')
```
